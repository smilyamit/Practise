Q26. Write a pandas program to create the Pivot table and find survival
rate by the gender, age of the different categories of various
classes.

```python
import pandas as pd
df = pd.read_csv('https://github.com/smilyamit/Practise/raw/master/pandas/titanic.csv')
age = pd.cut(df['age'], [0, 20, 50, 70])
piv_re = df.pivot_table('survived', index=['sex', age],columns='pclass')
piv_re
```

Q27. Write a pandas program to create the Pivot table and calculate
